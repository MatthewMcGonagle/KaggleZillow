```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get training property info

filename = 'Data/train_2016.csv'
trainid_df = pd.read_csv(filename)

# Get property info

filename = 'Data/properties_2016.csv'
properties_df = pd.read_csv(filename)

properties_df
```

```python
# Put respective properties_df info into train_df.

properties_df.set_index('parcelid', inplace = True)
trainid_df.set_index('parcelid', inplace = True)

train_df = trainid_df.join(properties_df)
```

```python
train_df
```

```python
train_df.columns
```

```python
# Bin Latitude and Longitude

nbins = 40
vars = ['latitude', 'longitude', 'logerror']
binvars = ['latitude', 'longitude', 'logerror']
vars_df = train_df[vars].dropna()
for i in range(len(binvars)):
    vals = vars_df[vars[i]].values
    
    vals = np.nan_to_num(vals)
    maxval = np.amax(vals)
    minval = np.amin(vals)
    vals = (vals - minval) / (maxval - minval) * nbins
    vals = vals.astype(int)
    vals = vals / nbins * (maxval - minval) + minval
    vars_df['bin' + str(i)] = vals
```

```python
# Look at statistics of log error depending on latitude and longitude.

grouped_df = vars_df.groupby(['bin0', 'bin1']).mean()

coordinates = grouped_df.index.values
coordinates = [[x,y] for x,y in coordinates]
coordinates = np.array(coordinates).T

vals = grouped_df['logerror'].values
minval = np.amin(vals)
maxval = np.amax(vals)
medianval = np.median(vals)
```

```python
# Get statistics on grouped logerror.

nbins = 40
vals = grouped_df['logerror'].values
minval = np.amin(vals)
maxval = np.amax(vals)
vals = (vals - minval) / (maxval - minval) * nbins
vals = vals.astype(int)
vals = vals / nbins * (maxval - minval) + minval
vals, counts = np.unique(vals, return_counts = True)

plt.plot(vals, counts)
plt.show()
```

```python
# Graph log error vs latitude and longitude for values of log error between minval and maxval.

vals = grouped_df['logerror'].values
minval = -0.05
maxval = 0.05

indices = (minval < vals) & (vals < maxval)
coordinates2 = (coordinates.T[indices]).T
vals2 = vals[indices]

# Plot range of interest

plt.scatter(coordinates2[0], coordinates2[1], c = vals2)
plt.gray()
plt.show()

coordinates2 = (coordinates.T[~indices]).T
vals2 = vals[~indices]

# Plot stuff not in range of interest

plt.scatter(coordinates2[0], coordinates2[1], c = vals2)
plt.gray()
plt.show()
```

```python
# Graph log error vs latitude and longitude for values of log error far from 0.
vals = grouped_df['logerror'].values
a = -0.1
b = 0.1
#vals = (vals - minval) / (maxval - minval)

indices = (a > vals) | (vals > b)
coordinates2 = (coordinates.T[indices]).T
vals2 = vals[indices]

# Plot range of interest

plt.scatter(coordinates2[0], coordinates2[1], c = vals2)
plt.gray()
plt.show()

coordinates2 = (coordinates.T[~indices]).T
vals2 = vals[~indices]

# Plot stuff not in range of interest

plt.scatter(coordinates2[0], coordinates2[1], c = vals2)
plt.gray()
plt.show()
```

```python
# Graph log error vs latitude and longitude for values of log error that is negative enough.
vals = grouped_df['logerror'].values
cutoff = -0.1

#vals = (vals - minval) / (maxval - minval)

indices = (cutoff > vals)
coordinates2 = (coordinates.T[indices]).T
vals2 = vals[indices]

# Plot range of interest

plt.scatter(coordinates2[0], coordinates2[1], c = vals2)
plt.gray()
plt.show()

coordinates2 = (coordinates.T[~indices]).T
vals2 = vals[~indices]

# Plot stuff not in range of interest

plt.scatter(coordinates2[0], coordinates2[1], c = vals2)
plt.gray()
plt.show()
```
